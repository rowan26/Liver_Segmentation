```python
# Cellule 1 — toujours en premier
import os
os.environ['KMP_DUPLICATE_LIB_OK'] = 'TRUE'
from monai.networks.nets import UNet
from monai.networks.layers import Norm
from monai.losses import DiceLoss, DiceCELoss

import torch
import nbimporter
from PreProcess_train import prepare
from Utilities import train

import matplotlib.pyplot as plt
data_dir = 'C:/Users/rowan/Desktop/Projets/Liver Segmentation/datasets/nifti_files'
model_dir = 'C:/Users/rowan/Desktop/Projets/Liver Segmentation/Results'

device=torch.device("cpu")
data_in=prepare(data_dir)
model = UNet(
    spatial_dims=3,
    in_channels=1,
    out_channels=2,
    channels=(16, 32, 64, 128, 256), 
    strides=(2, 2, 2, 2),
    num_res_units=2,
    norm=Norm.BATCH,
).to(device)

loss_function = DiceLoss(to_onehot_y=True, sigmoid=True, squared_pred=True)
optimizer = torch.optim.Adam(model.parameters(),1e-4, weight_decay=1e-4, amsgrad=True)
```

```python
train(model, data_in, loss_function, optimizer, 100, model_dir)
```
