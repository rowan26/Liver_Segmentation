## Create group of 75 slices

```python
in_path = 'C:/Users/rowan/Desktop/Projets/Liver Segmentation/datasets/DICOM_files/Labels'
out_path='C:/Users/rowan/Desktop/Projets/Liver Segmentation/datasets/DICOM_groups/labels'
```

```python
from glob import glob
import shutil
import os
import dicom2nifti
```

```python
for patient in glob(in_path + '/*'):
    patient_name= os.path.basename(os.path.normpath(patient))
    number_folders = int(len(glob(patient+'/*'))/75) #to len of patients / number of slices = number of folders needed
    
    #creation of subfolders with the number of slices DICOM (75 slices by folder)
    for i in range(number_folders) :
        output_path_name = os.path.join(out_path, patient_name + '_' + str(i))
        os.mkdir(output_path_name)
        for i,  file in enumerate(glob(patient+'/*')) :
            if i == 75 + 1 :
                break
            shutil.move(file, output_path_name)
```

## Convert the DICOM files into .nii NIftI files

## Find Empty (patient with any labels)

```python
import nibabel as nib
import numpy as np
```

```python
input_nitfi_file_path = 'C:/Users/rowan/Desktop/Projets/Liver Segmentation/datasets/nifti_files/labels/*'
images_path = 'C:/Users/rowan/Desktop/Projets/Liver Segmentation/datasets/nifti_files/images/'

list_labels_slices = glob(input_nitfi_file_path)

for patient in list_labels_slices:
    nifti_file = nib.load(patient)
    fdata = nifti_file.get_fdata()
    unique_datas = np.unique(fdata)
    
    if len(unique_datas) == 1:  # groupe vide
        filename_label = os.path.basename(patient)
        
        # liver_0_2_0.nii → on split par '_' → ['liver', '0', '2', '0.nii']
        # on retire l'index 1 (patient id) → liver_2_0.nii
        parts = filename_label.split('_')        # ['liver', '0', '2', '0.nii']
        filename_image = '_'.join([parts[0]] + parts[2:])  # liver_2_0.nii
        
        image_file = os.path.join(images_path, filename_image)
        
        os.remove(patient)      # supprime le label
        os.remove(image_file)   # supprime l'image
        
        print(f"Supprimé : {filename_label} + {filename_image}")
```
